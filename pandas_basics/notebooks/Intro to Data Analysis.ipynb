# Intro to Data Analysis
Businesses are now faced with a really good problem to have; too much data, and not enough resources to interpret it. 

We can now gather data that allows us to improve product recommendations and quickly identify areas of consumer interest and areas of our business that need improvement. In this growing world of big data, business leaders will be expected to understand how to harness this information to make sound business decisions. This course aims to give you the basic skills necessary to analyze and interpret a dataset. 

## What you'll learn
- How to use statistics to distinguish true correlation from noise
- How to interpret results to gain insight into which variables (your firm’s spending on advertising) might be related to or are driving other variables (sales)
- How to effectively report on and visualize these relationships using python and Jupyter Notebook

# Real Life Examples



## [Google Great Manager Awards](https://hbr.org/2013/12/how-google-sold-its-engineers-on-management)
Google, as we all know, is one of the most influential and rapidly growing tech companies on the planet. As it began it's journey of rapid growth, it decided to focus on the best ways to structure its management hierarchy. 

Google's People Analytics Department conducted a review of: 
- Employee reviews
- Employee satisfaction surveys
- Exit interview data

Using this information, they have successfully setup a management structure that works well with difficult-to-manage software engineers (yes, we certainly can be!) and identified 8 strategies of effective leaders that helped determine the right people to lead teams. 

### A good manager:
1. Is a good coach 
2. Empowers the team and does not micromanage (See the sidebar “How Google Defines One Key Behavior”) 
3. Expresses interest in and concern for team members’ success and personal well-being 
4. Is productive and results-oriented 
5. Is a good communicator—listens and shares information 
6. Helps with career development 
7. Has a clear vision and strategy for the team 
8. Has key technical skills that help him or her advise the team



## [WalMart Emergency Preparedness](https://www.nytimes.com/2004/11/14/business/yourmoney/what-walmart-knows-about-customers-habits.html)
A now-famous example of a company capitalzing on the results of data analysis. 

With the incoming threat of Hurricane Frances in 2004, WalMart had their team of analysts mine purchases made by customers prior to past catastrophes and used these results to tailor inventory towards customer's specific disaster needs (surprising beer and pop-tarts were very high on the list). This helped keep consumers happy and well-supplied during a disaster, and turned a huge profit for the company.
